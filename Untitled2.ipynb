```python
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split
from random import shuffle
from skimage import color
from skimage.transform import resize
from sklearn import svm
from skimage.feature import hog
from sklearn import metrics
from tqdm import tqdm
import matplotlib.pyplot as plt
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
from google.colab import auth
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
import pickle
auth.authenticate_user()
gauth = GoogleAuth()
drive = GoogleDrive(gauth)
local_download_path = os.path.expanduser('~/data')
try:
  os.makedirs(local_download_path)
except: pass
Benign_BreastTraining='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/benign/Train'
Benign_BreastTesting='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/benign/Test'
Malignant_BreastTraining='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/malignant/Train'
Malignant_BreastTesting='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/malignant/Test'
Normal_BreastTraining='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/normal/Train'
Normal_BreastTesting='/content/drive/MyDrive/CV2023CSYSDataset/Breast scans/normal/Test'

```

```python
#Preprocessing & Feature Extraction => TESTING DATA  

TestingX=[]
for img in tqdm(os.listdir(Benign_BreastTesting)): 
      path = os.path.join(Benign_BreastTesting, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data1 = cv2.resize(img_data, (64, 128))
      fdTest1, hog_Benign = hog(img_data1, orientations=16, pixels_per_cell=(8, 8),
               cells_per_block=(2, 2), visualize=True, multichannel=False)
      TestingX.append(fdTest1) 

for img in tqdm(os.listdir(Malignant_BreastTesting)): 
      path = os.path.join(Malignant_BreastTesting, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data2 = cv2.resize(img_data, (64, 128))
      fdTest2, hog_MAlignant = hog(img_data2, orientations=16, pixels_per_cell=(8, 8),
               cells_per_block=(2, 2), visualize=True, multichannel=False)    
      TestingX.append(fdTest2)

for img in tqdm(os.listdir(Normal_BreastTesting)): 
      path = os.path.join(Normal_BreastTesting, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data2 = cv2.resize(img_data, (64, 128))
      fdTest3, hog_Normal = hog(img_data2, orientations=16, pixels_per_cell=(8, 8),
               cells_per_block=(2, 2), visualize=True, multichannel=False)    
      TestingX.append(fdTest3)

TestData=np.array(TestingX)

#Preprocessing & Feature Extraction => TRAINING DATA  
TrainingX=[]
TrainingY=[]
label=[]

for img in tqdm(os.listdir(Benign_BreastTraining)): 
      path = os.path.join(Benign_BreastTraining, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data1 = cv2.resize(img_data, (64, 128))
      fd1, hog_Benigen = hog(img_data1, orientations=16, pixels_per_cell=(8, 8),
              cells_per_block=(2, 2), visualize=True, multichannel=False)
      TrainingX.append(fd1) 
      label.append(5)
for img in tqdm(os.listdir(Malignant_BreastTraining)): 
      path = os.path.join(Malignant_BreastTraining, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data2 = cv2.resize(img_data, (64, 128))
      fd2, hog_Malignant = hog(img_data2, orientations=16, pixels_per_cell=(8, 8),
               cells_per_block=(2, 2), visualize=True, multichannel=False)    
      TrainingX.append(fd2)
      label.append(4)

for img in tqdm(os.listdir(Normal_BreastTraining)): 
      path = os.path.join(Normal_BreastTraining, img)
      img_data = cv2.imread(path)
      img_data = color.rgb2gray(img_data)
      img_data2 = cv2.resize(img_data, (64, 128))
      fd3, hog_Normal = hog(img_data2, orientations=16, pixels_per_cell=(8, 8),
               cells_per_block=(2, 2), visualize=True, multichannel=False)    
      TrainingX.append(fd3)
      label.append(3)

TData=np.array(TrainingX)
X_train, X_test, y_train, y_test = train_test_split(TrainingX, label, test_size=0.2,random_state=100) 
BreastModel=svm.SVC(kernel='linear')
BreastModel.fit(X_train, y_train)
y_pred = BreastModel.predict(X_test)
N_pred=BreastModel.predict(X_train)

print("Testing Accuracy:",metrics.accuracy_score(y_test,y_pred))
print(" Training Accuracy:",metrics.accuracy_score(y_train,N_pred))

```
